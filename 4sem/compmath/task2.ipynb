## Основные теоретические положения

$\quad$**Группа** - это множество *G* с бинарной операцией *@*, удовлетворяющее следующим свойствам:
1. Бинарная операция *@* ассоциативна: $a@(b@c) = (a@b)@c$, для любых *a,b,c* из *G*
2. В множестве *G* существует нейтральный элемент: $a@e = e@a = a$, для любых *a* из *G*
3. Для любого элемента *a* из множества *G* сущесвует обратный элемент относительно операции *@*: *A* обратный *a* из *G* $a@A = A@a = e$

Также существует ряд "специальных типов" групп: **циклическая** - группа, порожденная одним элементом; **коммутативная** - группа с коммутативной бинарной операцией; про **симметрическую** группу следует сказать подробнее.

$\quad$Пусть *M = {1, 2, ..., n}* - конечное множество из *n* элементов. **Симметрической** называется группа $S_n = S(M)$ всех взаимно однозначных отображений *M* на себя с операцией умножения. Элементы симметрической группы называются **подстановками**.

$\quad$Подстановка называется **циклом**, если на некотором подмножестве множества *М* она действует следующим образом:
$$\begin{bmatrix}a_{i1} & a_{i2} & ... & a_{ik-1} & a_{ik} \\ a_{ik} & a_{i1} & ... & a_{ik-2} & a_{ik-1} \end{bmatrix}$$
, а остальные элементы неизменны. **Циклы** независимы, если действуют на непересекающихся подмножествах *М*.

$\quad$**Теорема**.
Каждая подстановка, не являющаяся нейтральным элементом, является произведением независимых циклов. Это разложение определено однозначно до порядка следования.

$\quad$**Следствие**.
Порядок подстановки равен НОК длин независимых циклов, входящих в разложение. Порядок подстановки = порядок циклической группы.

Пусть *G* - группа с нейтральным элементом *e*, а *X* - множество.
$\quad$**Действие группы *G* на множество *X*** происходит, если задана операция, обладающая для люого элемента *x* из *X* и для любых *g, h* из *G* следующими свойствами:
* $e@x = x$
* $gh(x) = g(h(x))$

$\quad$**Орбитой** *x* из *X* под действием *G* называется множество $G_x = \{gx : g \in G\}$. Количество элементов такого множества $G_x$ называется **длиной орбиты**.

$\quad$**Неподвижными точками** $g \in G$ называются точки *x* из *X*, для которых $gx = x$. Множество неподвижных точек обозначается $X^g$.

$\quad$**Лемма Бёрнсайда**. количество орбит действия *G* на *x*:
$N = {{1} \over {|G|}} \sum_{g \in G} |X^g|$.

Пусть *I* - произвольное множество, *C* - множество цветов. $\quad$**Раскраской множества *I*** называется функция из *I* в *C*. **Множество раскрасок** обозначается $C^I$. **Количество раскрасок**: $|C|^{|I|}$.

$\quad$**Лемма**. Пусть *G* - группа, действующая на *I*. Тогда $gf(i) = f(g^{-1}i)$ (где $g \in G, i \in I, f \in C^I$) задаёт действие *G* на $C^I$

$\quad$**Лемма**. Для $g \in G$ и $f \in C^I$ эквивалентны следующие условия:
1. $gf = f$
2. $f(i) = f(g^ni)$ для любых $i \in I$ и $n \in Z$
3. Если $i, j \in I$ лежат в одном цикле в циклической записи перестановки $\varphi(g)$, то $f(i) = f(j)$.

$\quad$**Следствие**. Количество раскрасок из $C^I$, которые сохраняют $g \in G$: $|C|^{C(\varphi (g))}$, где $C(\sigma)$ обозначает количество независимых циклов в записи перестановки $\sigma$.

$\quad$**Лемма**. Пусть $\sigma$ является циклом длины *к*, а *НОД*$(k,n) = d$, значит перестановка $\sigma^n$ является произведением *d* циклов $k \over d$.
